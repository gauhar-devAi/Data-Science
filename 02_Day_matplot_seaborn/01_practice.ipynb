```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Matplotlib ka basic tareeqah:
# Data
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y, color='green', marker='o') # Line plot
plt.title("My First Plot")                  # Title
plt.xlabel("X-Axis Label")                 # X label
plt.ylabel("Y-Axis Label")                 # Y label
plt.show()                                 # Graph display karne ke liye

```

```python
# Samajh lo 'df' aapka supermarket ka data hai
df = pd.read_csv("SuperMarket Analysis.csv")
```
