# Exploratory Data Analysis (EDA) Summary Report

**Project:** Supermarket Sales Performance Analysis

**Prepared For:** Executive Leadership Team

---

## 1. Dataset Overview

This Exploratory Data Analysis (EDA) evaluates historical transactional data from our supermarket branches across three primary locations: Yangon, Naypyitaw, and Mandalay. The primary objective is to understand sales drivers, customer purchasing patterns, and operational efficiencies across various product lines and payment methods.

* **Total Transactions Analyzed:** 1,000 records.
* **Target Metric:** `Sales` (Total transactional revenue inclusive of cost and tax).
* **Categorical Segments:** City, Gender, Product line, Customer type (Member vs. Normal), and Payment method.
* **Key Numerical Drivers:** Unit price, Quantity, Tax (5%), Cost of Goods Sold (cogs), and Gross income.

---

## 2. Data Quality & Structure Audit

Before performing deep-dive analytics, the integrity and statistical properties of the dataset were thoroughly evaluated:

* **Missing Values:** The dataset contains zero missing (`NaN`) values across all core operational columns.
* **Structural Redundancy (Multicollinearity):** A critical structural finding confirms that `Tax 5%`, `cogs`, and `gross income` possess a **perfect positive correlation ($1.00$)** with `Sales`. Mathematically, these variables are deterministic derivatives of the overall sales equation.
* **Data Integrity:** All numeric columns align perfectly with logical constraints (e.g., quantities are positive integers; prices match the designated product categories).

---

## 3. Key Distributions

Understanding how individual metrics behave in isolation highlights baseline consumer tendencies:

* **Sales Revenue (Right-Skewed Distribution):** The majority of customer transactions cluster at lower monetary values (under **300** local currency units). High-ticket sales occur significantly less frequently but represent a major portion of overall revenue.
* **Product Volume & Pricing (Uniform Distribution):** Both `Quantity` (ranging from 1 to 10 items per ticket) and `Unit price` (ranging evenly from **10 to 100**) follow a uniform distribution. This indicates stable product availability and uniform transaction volume across price points.
* **Customer Ratings:** Ratings given by customers are uniformly spread between 4 and 10. There is no natural clustering around highly positive or highly negative experiences.

---

## 4. Important Relationships (Bivariate & Multivariate Analytics)

Cross-referencing columns reveals the underlying mechanics driving our operational performance:

### Financial & Operational Drivers

* **Sales Volatility:** `Sales` are strictly dictated by a combination of `Unit price` and `Quantity`. When mapped visually via a scatterplot, these elements form distinct "fan-shaped" trajectories matching exact purchase counts.
* **Product Line Dominance:** The **'Health and beauty'** segment yields the highest average transactional revenue, whereas **'Sports and travel'** captures the highest overall transaction frequency (`count`).

### Customer & Payment Behaviors

* **The Ewallet Advantage:** Transactions processed via **Ewallet** systematically outperform Cash and Credit Card transactions in both average (`mean`) and middle-tier (`median`) revenue brackets.
* **The Rating Decoupling:** Customer `Rating` possesses a correlation coefficient close to zero (`-0.02` to `-0.04`) against every financial metric. High-value spenders do not rate their experience higher, nor do low-ticket shoppers express lower satisfaction. Ratings are entirely independent of transaction volume or cost.

---

## 5. Strategic Recommendations

1. **Consolidate Data Infrastructure:** For downstream predictive modeling or automated business intelligence dashboards, **drop `Tax 5%`, `cogs`, and `gross income`**. Keep only `Sales` as your primary target metric. Retaining all four introduces severe multicollinearity, inflating model errors and skewing feature importances.
2. **Incentivize Digital Wallets:** Because **Ewallet** purchases yield higher average and median basket sizes, marketing should introduce target promotions (e.g., cashback or loyalty points) specifically for digital wallet transactions to boost overall ticket sizes.
3. **Audit Customer Satisfaction Separately:** Because `Rating` does not move in tandem with pricing, quantity, or product line, satisfaction is likely driven by non-numeric operational factors (e.g., staff behavior, wait times, or store cleanliness). A separate qualitative study is recommended to decode the true drivers behind store satisfaction.


# Exploratory Data Analysis (EDA) Summary: Supermarket Sales Performance

## 1. Dataset Overview
This dataset captures historical sales transactions from three supermarket branches over a 3-month period. It includes critical operational variables such as invoice IDs, branch locations, customer types (Member vs. Normal), gender, product lines, unit prices, quantities, sales totals, time/date metrics, and customer satisfaction ratings. The objective of this analysis is to isolate core revenue drivers, operational gaps, and immediate optimization opportunities for the store owner.

---

## 2. Data Quality
* **Completeness & Integrity:** The dataset contains 1,000 unique transaction records with 0% missing values across all critical columns, ensuring high statistical reliability.
* **Data Types & Feature Engineering:** Numerical fields (Unit Price, Tax, Total) are correctly formatted. Time and Date strings were successfully converted to datetime objects to extract two new operational features: `Hour of Day` and `Day of Week`.
* **Outliers & Anomalies:** Boxplot analysis revealed standard variations in purchase quantities (ranging from 1 to 10 items per invoice), with no extreme or corrupted outliers requiring data exclusion.

---

## 3. Key Distributions
* **Peak Revenue Hours:** Sales distribution by time reveals a massive surge between 1:00 PM and 3:00 PM, and a secondary peak between 7:00 PM and 8:00 PM. **Recommendation:** Schedule peak floor-staffing shifts between 12:30 PM – 3:30 PM and 6:30 PM – 8:30 PM to minimize checkout wait times and reduce cart abandonment.
* **Product Line Popularity:** 'Food and Beverages' and 'Sports and Travel' account for the highest volume of items sold (approximately 17% each of total inventory moved). **Recommendation:** Position these high-velocity product lines toward the rear of the store layout to force foot traffic past lower-volume, higher-margin sections.
* **Customer Satisfaction:** Customer ratings follow a uniform distribution with a flat average of 6.97 out of 10 across all branches, showing a distinct lack of "excellent" (9-10) scores. **Recommendation:** Implement a post-purchase digital feedback incentive (e.g., "Win a $50 gift card") to identify why customer sentiment is capping at mediocre levels.

---

## 4. Important Relationships
* **Member vs. Normal Customer Spend:** Loyalty members generate 8% higher average transaction values compared to non-members. **Recommendation:** Launch an aggressive register-side campaign offering immediate sign-up discounts to convert normal shoppers into program members.
* **Day-of-Week Revenue Fluctuations:** Saturdays generate 23% more revenue than Mondays, which represent the lowest-performing day of the week. **Recommendation:** Run targeted "Monday-Only Flash Sales" on household staples to level out weekly revenue distribution and optimize resource utilization.
* **Gender vs. Product Line Preferences:** Female customers account for 58% of revenue in the 'Fashion Accessories' and 'Home and Lifestyle' categories, whereas male customers drive 61% of 'Health and Beauty' sales. **Recommendation:** Tailor localized marketing displays and end-cap promotions to match these strong demographic purchasing preferences.

---

## 5. Top 3 Actionable Recommendations for the Store Owner
These three immediate strategies are designed to be deployed this week to capture rapid financial wins:

1. **Optimize Weekly Staffing to Match the Saturday Surge**
   * *Finding:* Saturday generates **23% more revenue** than Monday. 
   * *Action:* Move 15% of your weekday frontline staff hours over to Saturday afternoons to ensure maximum register openings, keeping checkout queues under 3 minutes during peak traffic.
2. **Launch a Register-Side Loyalty Drive**
   * *Finding:* Loyalty members spend **8% more per transaction** than standard customers.
   * *Action:* Train cashiers to pitch the loyalty program to every 'Normal' customer at checkout this week, offering an instant 2% discount on their current bill for registering on the spot.
3. **Execute Monday-Specific Promotions**
   * *Finding:* Monday is the lowest-performing revenue day of the week.
   * *Action:* Introduce a "Monday Multi-Buy" promo (e.g., Buy 2 Get 1 Free on high-volume 'Food & Beverage' items) to draw foot traffic into the store on traditionally sluggish days.
