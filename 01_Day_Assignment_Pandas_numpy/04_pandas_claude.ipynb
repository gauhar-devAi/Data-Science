```python
import pandas as pd
import numpy as np
```

```python
# import pandas as pd — create a DataFrame from your products/prices data
# pd.DataFrame({'product': products, 'price': prices_arr, 'qty': qty, 'revenue': revenue})

# Original product and price data
products = [
    "Khaadi Lawn", "J. Kurta", "Peshawari Chappal", "Dalda Oil (5L)", 
    "National Biryani Masala", "Johar Joshanda (Box)", "Dawlance Fridge", 
    "Audionic Earbuds", "Sunsilk Shampoo", "Tapal Danedar"
]
prices = [3600, 4500, 2800, 2600, 130, 600, 85000, 32000, 580, 250]


# "We cannot perform direct mathematical operations (like multiplication and addition) 
# on an entire Python list all at once, whereas NumPy arrays make this process much easier and faster."
prices_arr = np.array(prices)

# Mock quantities for each product
qty = [12, 8, 15, 20, 50, 30, 3, 7, 25, 40]


# Calculate revenue (price * quantity)
revenue = prices_arr * qty

# Create Data Frame
df = pd.DataFrame({
    'product': products,
    'price': prices_arr,
    'qty': qty,
    'revenue': revenue
})

# Save the DataFrame to a CSV file
df.to_csv('products_data.csv')


# Is sary kaam sy hum ny apni 2 lists sy csv files bnai hain
```

```python
# Answer 3 questions using pandas: Which city has most sales? What is the average rating? Which product line earns most?
# df['City'].value_counts() is your starting point.

df = pd.read_csv('SuperMarket Analysis.csv')
# first 3 business questions

#City with the most sales (highest total revenue)
# it will sum every individual  country sale and sum ad show more sale above 
city_sales = df.groupby('City')['Sales'].sum().sort_values(ascending=False)
city_sales
```

```python
# What is the average rating
rating = df["Rating"].mean()
rating
```

```python
# Which product line earns most?
product_line = df.groupby('Product line')["Sales"].sum().sort_values(ascending=False)
product_line
```
