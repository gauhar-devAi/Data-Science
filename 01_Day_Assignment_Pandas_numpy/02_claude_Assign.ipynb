```python
market_items = {
    "Khaadi Lawn": 3699,
    "J. Kurta": 4500,
    "Peshawari Chappal": 2800,
    "Dalda Oil (5L)": 2650,
    "National Biryani Masala": 130,
    "Johar Joshanda (Box)": 600,
    "Dawlance Fridge": 85000,
    "Audionic Earbuds": 3200,
    "Sunsilk Shampoo": 580,
    "Tapal Danedar": 250
}

# Defining the function with a default threshold of 300 PKR
def classify_market_items(items_dict, threshold_limit=300):
     print(f"--- Classifying items with a threshold of {threshold_limit} PKR ---")

     for product, price in items_dict.items():
        # If/Else block to assign status
        if price <= threshold_limit:
            status = "affordable"
        else:
            status = "expensive"
        
        # Printing the formatted line
        print(f"{product} - {price} PKR - {status}")

         
# Running the function with your 300 PKR threshold
classify_market_items(market_items, 300)         
    

```

```python
# Converting our previous lists into a list of dictionaries
pakistan_market_dataset = [
    {"product": "Khaadi Lawn", "price": 3699},
    {"product": "J. Kurta", "price": 4500},
    {"product": "Peshawari Chappal", "price": 2800},
    {"product": "Dalda Oil (5L)", "price": 2650},
    {"product": "National Biryani Masala", "price": 130},
    {"product": "Johar Joshanda (Box)", "price": 600},
    {"product": "Haier 1.5 Ton AC", "price": 145000},
    {"product": "Dawlance Fridge", "price": 85000},
    {"product": "Audionic Earbuds", "price": 3200},
    {"product": "Sunsilk Shampoo", "price": 580}
]
```

```python
def get_expensive_from_dataset(dataset, limit):
    print(f"--- Items costing more than {limit:,} PKR ---")
    
    # Loop through each individual dictionary in our list
    for item in dataset:
        # Access values directly using the keys
        if item["price"] > limit:
            print(f"{item['product']}: Rs. {item['price']:,}")

# Running the function with a 500 PKR limit
get_expensive_from_dataset(pakistan_market_dataset, 500)
```
