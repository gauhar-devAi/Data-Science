```python
# Reshape a 1D array into a 2D array (rows × columns)
# np.arange(12).reshape(3, 4) — understanding shape is critical for ML later.

# np.arange(12) Kya Karta Hai?Yeh zero (0) se shuru kar ke 12 elements ki aik seedhi (1D) array banata hai:
# .reshape(3, 4) Kya Karta Hai?Yeh is seedhi line ko utha kar aik Matrix (2D array) mein badal deta hai, 
# jisme:3 Rows (lambi lines) hoti hain.4 Columns (khari lines) hote hain.
# Hamare paas 12 elements the.3 × 4 = 12 (Yeh bilkul sahi hai!)




```

```python
# 1. 0 se 11 tak ki 1D array banayi (Total 12 elements)
original_array = np.arange(12)
print("1D Array:")
print(original_array)
print("Original Shape:", original_array.shape) # Output: (12,)
print("-" * 40)

# 2. Isko 3 rows aur 4 columns mein reshape kiya
matrix_3x4 = original_array.reshape(3,4)
print("2D Array (3 Rows, 4 Columns):")
print(matrix_3x4)

print("New Shape:", matrix_3x4.shape) # Output: (3, 4)
print("-" * 40)

# 3. Aik aur example: 4 rows aur 3 columns
matrix_4x3 = original_array.reshape(4,3)
print("2D Array (4 Rows, 3 Columns):")
print(matrix_4x3)



```
