```python
import numpy as np
import pandas as pd
from IPython.display import Markdown, display

# 1. Load Dataset
df = pd.read_csv("SuperMarket Analysis.csv")

# 2. Drop Redundant Columns
df.drop(
    columns=[
        "Tax 5%",
        "cogs",
        "gross income",
        "Invoice ID",
        "gross margin percentage",
    ],
    inplace=True,
)

# 3. Binary Label Encoding
df["Gender_encoded"] = (df["Gender"] == "Female").astype(int)
df["Member_encoded"] = (df["Customer type"] == "Member").astype(int)

# 4. One-Hot Encoding (Dummy Variable Trap avoided)
df = pd.get_dummies(
    df, columns=["City", "Product line", "Payment"], drop_first=True
)


# 5. Feature Extraction: Hour and Day of Week (Fixed with format='mixed')


# pd.to_datetime(): Pandas ka yeh function string ya text format mein likhe hue time/date ko aik special format
# (datetime64) mein convert karta hai, taake computer ko pata chal sake ke yeh koi aam text nahi balkay waqt aur
# tareekh ka record hai.

# format='%H:%M' aur format='%m/%d/%Y': Hum computer ko guide kar rahe hain ke hamare data mein ghante, mint aur
# saal kis tarteeb se likhe hue hain taake parsing mein koi error na aaye.

# .dt.hour: .dt Pandas ka datetime accessor hai. Jab hum iske aage .hour lagate hain, to yeh pure time se sirf 
# ghanta (jaise 13:45 se 13) nikal kar naye column mein daal deta hai.

# .dt.dayofweek: Yeh tareekh ko check karta hai aur us din ke hisab se aik number assign kar deta hai. Pandas 
# mein standard indexing hoti hai:
# 0 = Monday


df["Hour"] = pd.to_datetime(df["Time"], format="mixed").dt.hour
df["DayOfWeek"] = pd.to_datetime(df["Date"], format="mixed").dt.dayofweek

# 6. Verification & Summary
shape_info = f"Rows: {df.shape[0]}, Columns: {df.shape[1]}"
unique_hours = sorted(df["Hour"].unique())
unique_days = sorted(df["DayOfWeek"].unique())

display(
    Markdown(
        f"""
### 1. Data Feature Extraction Summary
* **Dataset Scale:** Naye features extract karne ke baad ab table mein **{df.shape[0]} rows** aur **{df.shape[1]} columns** hain.
* **Feature Extraction Complete:** * `Time` column se `Hour` extract kiya gaya (Unique hours detected: {unique_hours}).
    * `Date` column se `DayOfWeek` extract kiya gaya (0 = Monday, 6 = Sunday | Range detected: {unique_days}).
* **Data Readiness:** Aapka data ab clean, encoded, aur errors se paak ho kar ready hai!
"""
    )
)
```
